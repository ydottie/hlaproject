```python
import pandas as pd
import numpy as np 
```

```python
df=pd.read_csv("3_gs.csv")
df=df[['Run','DRB1']]
df
```

```python
df[['DRB1.1','DRB1.2']]=df.DRB1.str.split('/',expand=True)
df=df.drop(['DRB1'], axis=1)
df=df.rename(columns={"DRB1.1": "DRB1", "DRB1.2": "DRB1.1"})
df
```

```python
for col in df.columns[df.columns!='Run']:
    if '.' in col:
        allele = col.split('.')[0]
        for i in range(len(df[col])):
            df[col][i]=allele+ df[col][i]
            df[col][i]=df[col][i].replace(" ", "")
    else: 
        for i in range(len(df[col])):
            df[col][i]=col+ df[col][i] 
            df[col][i]=df[col][i].replace(" ", "")
```
