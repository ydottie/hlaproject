```python
import pandas as pd
import glob
```

```python
loc = '/content/gdrive/MyDrive/test/MASTER RESULTS DIRECTORY'
df = pd.DataFrame()
```

```python
with open(loc+'/accession/d2_list.txt') as f:
  accessionnumbers = f.readlines()
```

```python
ERR,A1,A2,B1,B2,C1,C2 = list(), list(), list(), list(), list(), list(), list()
```

```python
for number in accessionnumbers: 
  print (number)
  # creating the path to the sample
  formatted_num = number.rstrip('\n')
  path = f"{loc}/raw/seq2HLA_d2/{formatted_num}*.HLAgenotype4digits"

  classIpath = glob.glob(path)
  s_path=("".join(classIpath))   

  # construct and initialize sample list
  sample = list()
  sample.append(formatted_num)

  # add all raw data to sample
  try:
    with open(s_path) as f:
      next(f)
      rawcontents = f.read()
      words = rawcontents.split('\t')

      for word in words:
        sample.append(word)

    # create separate lists per gene
    if (len(sample) > 1):
      ERR.append(sample[0])
      A1.append(sample[2])
      A2.append(sample[4])
      B1.append(sample[6])
      B2.append(sample[8])
      C1.append(sample[10])
      C2.append(sample[12])

  except FileNotFoundError:
    pass
```

```python
dfERR = pd.DataFrame(ERR)
dfA1 = pd.DataFrame(A1)
dfA2 = pd.DataFrame(A2)
dfB1 = pd.DataFrame(B1)
dfB2 = pd.DataFrame(B2)
dfC1 = pd.DataFrame(C1)
dfC2 = pd.DataFrame(C2)
first_csv = pd.concat([dfERR,dfA1,dfA2,dfB1,dfB2,dfC1,dfC2], axis=1, join='inner')
first_csv
```

```python
first_csv.to_csv('seq2hla_d2.csv', index=False)
! cp seq2hla_d2.csv "/content/gdrive/MyDrive/test/MASTER RESULTS DIRECTORY/standard"
```
